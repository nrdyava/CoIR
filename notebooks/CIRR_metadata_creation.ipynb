```python
import os
import json
```

## Corpus Data

### Train

```python
train_corpus = json.load(open('/proj/vondrick4/naveen/coir-data/CIRR/image_splits/split.rc2.test1.json', 'r'))
```

```python
train_corpus
```

```python
train_corpus_out = []
for img_id, img_path in train_corpus.items():
    temp = {}
    temp['image-id'] = img_id
    temp['image-path'] = os.path.join('img_raw', img_path[2:])
    train_corpus_out.append(temp)
```

```python
len(train_corpus_out)
```

```python
with open('/proj/vondrick4/naveen/coir-data/CIRR/metadata/test1_corpus.json', 'w') as json_file:
    json.dump(train_corpus_out , json_file, indent=4)
```
