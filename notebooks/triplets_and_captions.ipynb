```python
coco_imgs_and_caps_train = json.load(open('/proj/vondrick4/naveen/coir-data/annotations/images_and_caps_train2014.json', 'r'))
```

```python
coco_imgs_and_caps_train_imgs = list(set(map(lambda x: x['image-id'], coco_imgs_and_caps_train)))
```

```python
coco_imgs_and_caps_train_imgs[:5]
```

```python
len(coco_imgs_and_caps_train_imgs)
```

```python
len(coco_imgs_and_caps_train)
```

```python
coco_imgs_and_caps_train_processed = defaultdict(list)
for sample in coco_imgs_and_caps_train:
    coco_imgs_and_caps_train_processed[sample['image-id']].append(sample['caption'])
```

```python
coco_imgs_and_caps_train_processed
```

```python
len(list(coco_imgs_and_caps_train_processed.keys()))
```

```python
with open("/proj/vondrick4/naveen/coir-data/annotations/images_and_caps_train2014_processed_dict.json", "w") as json_file:
    json.dump(coco_imgs_and_caps_train_processed, json_file)
```

```python
temp = 0
for key, value in coco_imgs_and_caps_train_processed.items():
    temp+=len(value)
```

```python
temp/len(coco_imgs_and_caps_train_processed.keys())
```
