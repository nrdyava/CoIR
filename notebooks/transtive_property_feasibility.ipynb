```python
import os
import json
from collections import defaultdict
```

```python
lasco_train_indexed = json.load(open('/proj/vondrick4/naveen/coir-data/LaSCo/metadata/lasco_train_indexed.json', 'r'))
```

```python
source_target_map = defaultdict(list)
for sample in lasco_train_indexed:
    source_target_map[sample['query-image-id']].append(sample['target-image-id'])
```

```python
source_target_map
```

```python
st_cnts = []
for k, v in source_target_map.items():
    st_cnts.append(len(v))
```
